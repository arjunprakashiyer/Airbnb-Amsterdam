# Mixed data type

```python
#Logic for checking mixed types in data frames
for col in df_listings.columns.tolist():
    weird = (df_listings[[col]].applymap(type) != df_listings[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_listings[weird]) > 0:
        print (col)
```

```python
# Changing the listing name, last_review and the host name to String data type
```

```python
df_listings['host_name'] = df_listings['host_name'].astype('str')
```

```python
df_listings['last_review'] = df_listings['last_review'].astype('str')
```
