### Standardizing the data

```python
# Use Standard Scaler to standardize data in the dataframe

scaler = StandardScaler()
segmentation_std = scaler.fit_transform(df_listings_kmeans)
```

```python
# Fit standardized data using PCA

pca_data = PCA()
pca_data.fit(segmentation_std)
```

```python
# Show how much variance is explained by each of the seven components

pca_data.explained_variance_ratio_
```

```python
plt.figure(figsize = (10,8))
plt.plot(range(1,9), pca_data.explained_variance_ratio_.cumsum(), marker = 'o', linestyle = '--')
plt.title('Explained Variance by Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
```

After seeing the above chart, we can see that the value for number of components we can use is 6 which corresponds to 80% of the variance(cumulative explained variance of 0.8)

```python
# Choice of six components based on the above findings

pca_data = PCA(n_components = 5)
```

```python
# Fit model to data

pca_data.fit(segmentation_std)
```

```python
# Extract calculated resulting components for our data set

pca_data.transform(segmentation_std)
```

```python
#Transforming the above pca data
scores_pca_final  = pca_data.transform(segmentation_std)
```

# Elbow Technique

```python
# Defines the range of potential clusters in the data.
num_cl = range(1, 10) 
kmeans = [KMeans(n_clusters=i) for i in num_cl] # Defines k-means clusters in the range assigned above.
```

```python
score = [kmeans[i].fit(scores_pca_final).score(scores_pca_final) for i in range(len(kmeans))] # Creates a score that represents 
# a rate of variation for the given cluster option.

score
```

```python
# Plot the elbow curve using PyLab.

pl.plot(num_cl,score, marker = 'o')
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

The ideal number of Clusters would be 3. The curve becomes smoother after the values of number of clusters- 3.

# k-Means Clustering

```python
# Create the k-means object.

kmeans = KMeans(n_clusters = 3, n_init=10) 
```

```python
# Fit the k-means object to the data.

kmeans.fit(scores_pca_final)
```

```python
#Running the Predict method
df_listings_kmeans['clusters'] = kmeans.fit_predict(df_listings_kmeans)
```

```python
#Checking the new Column
df_listings_kmeans.head()
```

```python
df_listings_kmeans['clusters'].value_counts()
```

```python
# Plot the clusters for the "Price" and "Availability_365" variables.

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df_listings_kmeans['price'], y=df_listings_kmeans['availability_365'], hue=kmeans.labels_, s=100) 
# Here, you're subsetting `X` for the x and y arguments to avoid using their labels. 
# `hue` takes the value of the attribute `kmeans.labels_`, which is the result of running the k-means algorithm.
# `s` represents the size of the points you want to see in the plot.

ax.grid(False) # This removes the grid from the background.
plt.xlabel('Price') # Label x-axis.
plt.ylabel('Availability of the listings for the Calendar Year') # Label y-axis.
plt.show()
```

```python
# Plot the clusters for the "House Age" and "Price per Unit" variables.

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df_listings_kmeans['number_of_reviews'], y=df_listings_kmeans['availability_365'], hue=kmeans.labels_, s=100)

ax.grid(False) 
plt.xlabel('Number of Reviews') 
plt.ylabel('Availability of the listings for the Calendar Year') 
plt.show()
```

```python
df_listings_kmeans.loc[df_listings_kmeans['clusters'] == 2, 'cluster'] = 'dark purple'
df_listings_kmeans.loc[df_listings_kmeans['clusters'] == 1, 'cluster'] = 'purple'
df_listings_kmeans.loc[df_listings_kmeans['clusters'] == 0, 'cluster'] = 'light pink'
```

```python
df_listings_kmeans.groupby('cluster').agg({'price':['mean', 'median'], 
                         'number_of_reviews':['mean', 'median'], 
                         'availability_365':['mean', 'median']
                          })
```
